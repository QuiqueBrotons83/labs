```python
# Lo primero que hacemos es importar las librerias Panda y Numpy. Y hacemos que se muuestren los anchos de columna.
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames

```

```python
datos_netflix_filtros = pd.read_csv("datos/datos_netflix.csv", index_col = 0)
datos_netflix_filtros.head(3)
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
titulos_2021 = datos_netflix_filtros[(datos_netflix_filtros['release_year'] == 2021) & (datos_netflix_filtros['type'] == 'Movie')]
titulos_2021
```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
duracion_noventa = datos_netflix_filtros[datos_netflix_filtros['duration'] == '90 min']
duracion_noventa[['Title', 'duration']]
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
india_lanzados = datos_netflix_filtros[(datos_netflix_filtros['country'] == 'India') & (datos_netflix_filtros['release_year'] == 2018)]
india_lanzados
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
clasificacion = datos_netflix_filtros[datos_netflix_filtros['rating'].isin(['TV-14', 'R'])]
clasificacion['Title']
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
directores =datos_netflix_filtros[datos_netflix_filtros['director'].isin(['Christopher Nolan', 'Martin Scorsese'])]
directores
```

```python
directores['director'].value_counts()
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python
titulos_paises = datos_netflix_filtros[datos_netflix_filtros['country'].isin(['United States','United Kingdom', 'Canada'])]
titulos_paises[['Title', 'country', 'Language']]
```

```python
titulos_lenguaje = titulos_paises[titulos_paises['Language'].str.contains(',', na=False)]
titulos_lenguaje
# No hay más de un lenguaje en estos titulos.
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python
datos_netflix_filtros[datos_netflix_filtros['release_year'].between(2015, 2020, inclusive='both')]
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python
datos_netflix_filtros[datos_netflix_filtros['IMDB Score'].between(6.5, 8.0, inclusive='both')]
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 y 150 minutos?

```python
datos_netflix_filtros[datos_netflix_filtros['duration'].isin(['90 min', '150 min'])]
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
titulos_love = datos_netflix_filtros[(datos_netflix_filtros['Title'].str.contains('Love', case=False,na=False)) & (datos_netflix_filtros['type'] == 'TV Show')]
titulos_love
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python
director_apellido = datos_netflix_filtros[(datos_netflix_filtros['director'].str.contains('Johnson', case=False,na=False))]
director_apellido[['Title', 'director']]
```

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?
