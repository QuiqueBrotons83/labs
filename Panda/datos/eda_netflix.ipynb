### Parte 3: Uso de `loc` e `iloc`


   - Selecciona todas las filas donde el tipo de contenido sea “Movie”.


```python
movies_datos_netflix = datos_netflix.loc[datos_netflix['type'] == 'Movie']
movies_datos_netflix
```

   - Muestra solo los títulos (title) y la duración (duration) de todos los contenidos que sean de tipo “TV Show”.


```python
titulo_duracion_netflix = datos_netflix.loc[datos_netflix['type'] == 'TV Show', ['Title', 'duration']]
titulo_duracion_netflix
```

   - Selecciona todas las filas donde el país sea “United States”.


```python
filas_United_States = datos_netflix.loc[datos_netflix['country'] == 'United States']
filas_United_States
```

   - Muestra todas las películas (type = “Movie”) y selecciona las columnas title y director.


```python
tipo_movie = datos_netflix.loc[datos_netflix['type'] == 'Movie', ['Title', 'duration']]
tipo_movie
```

   - Selecciona los títulos (title) y géneros (listed_in) de todas las producciones lanzadas en 2018.


```python
producciones_2018 = datos_netflix.loc[datos_netflix['release_year'] == 2018, ['Title', 'listed_in']]
producciones_2018
```

   - Selecciona las filas donde la columna director esté vacía y muestra solo los títulos (title).


```python
director_desconocido = datos_netflix.loc[datos_netflix['director'] == 'Desconocido', ['Title']]
director_desconocido
```

2. **Acceso a datos específicos con `iloc`:**


   - Selecciona las primeras 5 filas del conjunto de datos y muestra solo las columnas title, director y country.


```python
cinco_filas = datos_netflix.iloc[:5,[datos_netflix.columns.get_loc('Title'), datos_netflix.columns.get_loc('director'), datos_netflix.columns.get_loc ('country')]]
cinco_filas
```

   - Muestra las últimas 5 filas del conjunto de datos y selecciona las columnas title y description.



```python
ultimas_filas = datos_netflix.iloc[-5:,[datos_netflix.columns.get_loc('Title'), datos_netflix.columns.get_loc('description')]]
ultimas_filas
```

   - Selecciona las primeras 10 filas del dataset y muestra las columnas desde la segunda hasta la quinta (incluidas). 


```python
primeras_diez = datos_netflix.iloc[:10, 1:6]
primeras_diez
```

   - Muestra las últimas 7 filas del dataset y selecciona las columnas title, director, y country usando índices posicionales.


```python
ultimas_siete = datos_netflix.iloc[-7:,[datos_netflix.columns.get_loc('Title'), datos_netflix.columns.get_loc('director'), datos_netflix.columns.get_loc('country')]]
ultimas_siete
```

   - Selecciona las filas desde la 15 hasta la 25 (incluidas) y muestra las primeras 4 columnas.
