```python
# importamos las librerías que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd
import numpy as np

# Para guardar DataFrames en Excel
# -----------------------------------------------------------------------
from pandas import ExcelWriter

# Para generar todas las posibles combinaciones
# -----------------------------------------------------------------------
import itertools

# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
# cargamos el dataframe creado en la lección anterior
df = pd.read_csv("datos/datos_netflix.csv", index_col = 0)
df
```

#### Ejercicio 1: ¿Cuántos títulos hay por cada tipo de contenido ('Movie' o 'TV Show')?

```python
df.groupby("type").count()
titulos_count = df.groupby("type")["Title"].count()
df_titulos_count = titulos_count.reset_index()
df_titulos_count
```

#### Ejercicio 2: ¿Cuál es el promedio de puntuación de las películas agrupadas por clasificación de audiencia ('rating')? Interpreta los resultados, ¿hay algo que te llame la atención? Explica por qué. 

```python
promedio_puntacion = df.groupby("Genre")["IMDB Score"].mean()
promedio_puntacion
df_promedio_puntuacion = promedio_puntacion.reset_index()
df_promedio_puntuacion
```

#### Ejercicio 3: ¿Cuántos títulos hay por cada país? Ordena los resultados de mayor a menor

```python
titulos_pais = df.groupby("country")["Title"].count().sort_values(ascending=False)
titulos_pais
df_titulos_pais = titulos_pais.reset_index()
df_titulos_pais
```

#### Ejercicio 4: ¿Cuál es la calificación promedio en IMDb de los títulos agrupados por tipo ('Movie' o 'TV Show')?

```python
calif_prom = df.groupby("type")["IMDB Score"].mean()
df_calif_prom = calif_prom.reset_index()
df_calif_prom
```

#### Ejercicio 5: ¿Cuántos títulos se añadieron a Netflix cada año? Para resolver este ejercicio tendrás uqe convertir la columna `date_added` a formato fecha usando el método `pd.to_datetime`. [Aquí](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html) tienes la documentación, leeala con detenimiento antes de usar el método. 

```python

df["date_added"] = pd.to_datetime(df["date_added"], errors='coerce')

df["separar_años"] = df["date_added"].dt.year

titulos_ano = df.groupby("separar_años")["Title"].count()
df_titulos_ano = titulos_ano.reset_index()
df_titulos_ano
```

#### Ejercicio 6: ¿Cuál es el país con más títulos lanzados en el año 2020? Para resolver este ejercicio después del gropuby os puede resultar interesante el método [`idxmax`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.idxmax.html)

```python
titulos_2020 = df[df["release_year"] == 2020]
mas_titulos = titulos_2020.groupby("country").size()
pais_mas_titulos = mas_titulos.idxmax()
print(f"Pais con más titulos en el 2020 es:{pais_mas_titulos}")
```

#### Ejercicio 7:¿Cuál es el promedio de puntuación en IMDb de los títulos agrupados por país?

```python
df["IIMDB Score"] = pd.to_numeric(df["IMDB Score"], errors="coerce")

puntuacion_promedio = df.groupby("country")["IMDB Score"].mean()
df_puntuacion_promedio = puntuacion_promedio.reset_index()
df_puntuacion_promedio
```

#### Ejercicio 8: ¿Cuántos títulos fueron lanzados en cada año agrupados por tipo ('Movie' o 'TV Show')?

```python
titulos_ano_1 = df.groupby(["release_year","type"]).size().reset_index(name='count')
df_titulos_ano_1
```

#### Ejercicio 9: ¿Cuál es el país con la mayor cantidad de títulos de clasificación 'TV-MA'?

```python
filtro_TVMa = df[df['rating'] == 'TV-MA']
titulos_clasi_pais = filtro_TVMa.groupby("country").size()
titulos_clasi_pais = titulos_clasi_pais.idxmax()
print(f"pais con más titulos: {titulos_clasi_pais}") 
```

#### Ejercicio 10. ¿Cuál es el promedio de duración de las películas agrupadas por país?

```python
películas = df[df['type'] == 'Movie']
# Esto tuve que indagar por internet
películas['duration'] = películas['duration'].str.extract('(\d+)').astype(float)

promedio_duración_por_país = películas.groupby('country')['duration'].mean()
df_promedio_duracion_por_pais = promedio_duración_por_país.reset_index()
df_promedio_duracion_por_pais
```

#### Ejercicio 11: ¿Cuál es el número total de títulos por país y clasificación de audiencia ('rating')?

```python
titulos_pais_rating = df.groupby(['country', 'rating']).size().reset_index(name='count')
titulos_pais_rating
```

#### Ejercicio 12: ¿Cuántos títulos fueron lanzados antes del año 2000 agrupados por país?

```python
antes_2000 = df[df['release_year'] < 2000]
titulos_antes_2000 = antes_2000.groupby('country').size().reset_index(name='count')
titulos_antes_2000.sample(10)
```

#### Ejercicio 13: ¿Cuál es la puntuación promedio en IMDb de los títulos lanzados en 2021 agrupados por país?

```python
titulos_2021 = df[df['release_year'] == 2021]
promedio_IMDB = titulos_2021.groupby('country')['IMDB Score'].mean()
promedio_IMDB.head(50)
```

#### Ejercicio 14: ¿Cuántos títulos fueron dirigidos por cada director? Ordena los resultados de mayor a menor

```python
titulos_director = df.groupby('director').size().reset_index(name='count')
titulos_director = titulos_director.sort_values(by='count', ascending=False)
titulos_director
```

#### Ejercicio 15: ¿Cuál es el lenguaje más frecuente en los títulos lanzados en los últimos 5 años agrupado por país?

```python
año_actual = pd.to_datetime('today').year

ultimos_5_años = df[df['release_year'] >= (año_actual- 5)]

lenguaje_pais = ultimos_5_años.groupby(['country', 'Language']).size().reset_index(name='count')
lenguaje_mas_frecuente = lenguaje_pais.loc[lenguaje_pais.groupby('country')['count'].idxmax()].sort_values(by='count', ascending=False)
lenguaje_mas_frecuente.head(10)

```
